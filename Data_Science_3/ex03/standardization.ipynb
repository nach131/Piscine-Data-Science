```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
dataTrain = pd.read_csv("subject/Train_knight.csv")
dataTrain.head()
```

```python
dataTest = pd.read_csv("subject/Test_knight.csv")
dataTest.head()
```

```python
# Escalamiento de datos
scaler = StandardScaler()
data = dataTrain[dataTrain.columns[:-1]]
knightTrain = dataTrain['knight']

data_scaled = scaler.fit_transform(data)
dataTrain_scaled = pd.DataFrame(data_scaled, columns=dataTrain.columns[:-1])
dataTrain_scaled = dataTrain_scaled.join(knightTrain)

# Agrupando los datos por 'knight'
grouped_data = dataTrain_scaled.groupby('knight')

# CreaciÃ³n del grÃ¡fico mejorado
fig, ax = plt.subplots(figsize=(10, 8))

# GrÃ¡fica para los Sith
ax.scatter(grouped_data.get_group('Sith')['Empowered'], 
           grouped_data.get_group('Sith')['Friendship'], 
           label='Sith', color='red', alpha=0.7, edgecolor='k', s=100)

# GrÃ¡fica para los Jedi
ax.scatter(grouped_data.get_group('Jedi')['Empowered'], 
           grouped_data.get_group('Jedi')['Friendship'], 
           label='Jedi', color='blue', alpha=0.7, edgecolor='k', s=100)

# Ajustes de estilo
ax.set_xlabel('Empowered', fontsize=14)
ax.set_ylabel('Friendship', fontsize=14)
ax.set_title('Empowered vs Friendship for Sith and Jedi', fontsize=16)
ax.legend(fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)

# Mostrar el grÃ¡fico
plt.show()
```

**Standardization** (EstandarizaciÃ³n) es tÃ©cnicas comunes de preprocesamiento de datos.


### **Standardization (EstandarizaciÃ³n)**
La **estandarizaciÃ³n** ajusta los datos de manera que cada caracterÃ­stica (columna) tenga una **media de 0** y una **desviaciÃ³n estÃ¡ndar de 1**. Esto significa que los valores se distribuyen alrededor de cero, y las desviaciones respecto de la media estÃ¡n normalizadas.

FÃ³rmula de la estandarizaciÃ³n:

$$Z = \frac{X - \mu}{\sigma}$$

Donde:
- $X$ es el valor original de la caracterÃ­stica.
- $\mu$ es la media de la caracterÃ­stica.
- $\sigma$ es la desviaciÃ³n estÃ¡ndar de la caracterÃ­stica.


### Ejemplo:
Si tienes una columna de datos que tiene un rango de valores muy grande, la estandarizaciÃ³n ajustarÃ¡ estos datos para que sigan una **distribuciÃ³n normal estÃ¡ndar**, donde:
- Media ($\mu$) es 0.
- DesviaciÃ³n estÃ¡ndar ($\sigma$) es 1.

**CuÃ¡ndo usar estandarizaciÃ³n:**
- Cuando los datos siguen una distribuciÃ³n normal o gaussiana (aproximadamente).
- Modelos que asumen normalidad en los datos, como la regresiÃ³n lineal o redes neuronales.
- Cuando los algoritmos requieren mÃ©tricas de distancia (KNN, SVM, redes neuronales).
