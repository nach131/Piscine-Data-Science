```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
dataTrain = pd.read_csv("subject/Train_knight.csv")
dataTrain.head()
```

```python
dataTest = pd.read_csv("subject/Test_knight.csv")
dataTest.head()
```

```python
# Escalamiento de datos
scaler = StandardScaler()
data = dataTrain[dataTrain.columns[:-1]]
knightTrain = dataTrain['knight']

data_scaled = scaler.fit_transform(data)
dataTrain_scaled = pd.DataFrame(data_scaled, columns=dataTrain.columns[:-1])
dataTrain_scaled = dataTrain_scaled.join(knightTrain)

# Agrupando los datos por 'knight'
grouped_data = dataTrain_scaled.groupby('knight')

# Creación del gráfico mejorado
fig, ax = plt.subplots(figsize=(10, 8))

# Gráfica para los Sith
ax.scatter(grouped_data.get_group('Sith')['Empowered'], 
           grouped_data.get_group('Sith')['Friendship'], 
           label='Sith', color='red', alpha=0.7, edgecolor='k', s=100)

# Gráfica para los Jedi
ax.scatter(grouped_data.get_group('Jedi')['Empowered'], 
           grouped_data.get_group('Jedi')['Friendship'], 
           label='Jedi', color='blue', alpha=0.7, edgecolor='k', s=100)

# Ajustes de estilo
ax.set_xlabel('Empowered', fontsize=14)
ax.set_ylabel('Friendship', fontsize=14)
ax.set_title('Empowered vs Friendship for Sith and Jedi', fontsize=16)
ax.legend(fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gráfico
plt.show()
```

**Standardization** (Estandarización) es técnicas comunes de preprocesamiento de datos.


### **Standardization (Estandarización)**
La **estandarización** ajusta los datos de manera que cada característica (columna) tenga una **media de 0** y una **desviación estándar de 1**. Esto significa que los valores se distribuyen alrededor de cero, y las desviaciones respecto de la media están normalizadas.

Fórmula de la estandarización:

$$Z = \frac{X - \mu}{\sigma}$$

Donde:
- $X$ es el valor original de la característica.
- $\mu$ es la media de la característica.
- $\sigma$ es la desviación estándar de la característica.


### Ejemplo:
Si tienes una columna de datos que tiene un rango de valores muy grande, la estandarización ajustará estos datos para que sigan una **distribución normal estándar**, donde:
- Media ($\mu$) es 0.
- Desviación estándar ($\sigma$) es 1.

**Cuándo usar estandarización:**
- Cuando los datos siguen una distribución normal o gaussiana (aproximadamente).
- Modelos que asumen normalidad en los datos, como la regresión lineal o redes neuronales.
- Cuando los algoritmos requieren métricas de distancia (KNN, SVM, redes neuronales).
