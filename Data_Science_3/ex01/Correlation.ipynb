````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data = pd.read_csv("subject/Train_knight.csv")
data.head()
````

````python
data = data.replace({'knight': 'Sith'}, False) # Convierte Sith por False
data = data.replace({'knight': 'Jedi'}, True) # Convierte Jedi por True
data.head()
````

````python
# Calcula la correlación
data.corr()['knight'].sort_values(ascending=False) 
````

## Cálculo de la correlación:

   ```python
   data.corr()['knight'].sort_values(ascending=False)
   ```
   - El método `.corr()` calcula la **matriz de correlación** de todas las columnas numéricas del `DataFrame` `data`.
   - La correlación mide la relación lineal entre dos variables, y su valor oscila entre -1 y 1:
     - **1** indica una correlación positiva perfecta (cuando una variable sube, la otra también sube).
     - **0** indica que no hay ninguna relación lineal entre las variables.
     - **-1** indica una correlación negativa perfecta (cuando una variable sube, la otra baja).
   - El código `.corr()['knight']` selecciona específicamente la columna de correlaciones con respecto a `knight`, es decir, calcula cómo se relaciona la columna `knight` con todas las demás columnas del `DataFrame`.
   - Finalmente, `.sort_values(ascending=False)` ordena estas correlaciones en orden **descendente**, mostrando primero las variables más correlacionadas positivamente con `knight`.

### ¿Por qué convertir los valores a `1` y `0`?

La conversión de etiquetas categóricas como 'Jedi' y 'Sith' a `1` y `0` se realiza para que se puedan aplicar operaciones matemáticas y estadísticas, como la correlación. La correlación solo puede calcularse entre columnas numéricas, por lo que este paso de preprocesamiento es necesario para realizar este tipo de análisis.
