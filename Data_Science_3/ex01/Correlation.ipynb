````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data = pd.read_csv("subject/Train_knight.csv")
data.head()
````

````python
data = data.replace({'knight': 'Sith'}, False) # Convierte Sith por False
data = data.replace({'knight': 'Jedi'}, True) # Convierte Jedi por True
data.head()
````

````python
# Calcula la correlaciÃ³n
data.corr()['knight'].sort_values(ascending=False) 
````

## CÃ¡lculo de la correlaciÃ³n:

   ```python
   data.corr()['knight'].sort_values(ascending=False)
   ```
   - El mÃ©todo `.corr()` calcula la **matriz de correlaciÃ³n** de todas las columnas numÃ©ricas del `DataFrame` `data`.
   - La correlaciÃ³n mide la relaciÃ³n lineal entre dos variables, y su valor oscila entre -1 y 1:
     - **1** indica una correlaciÃ³n positiva perfecta (cuando una variable sube, la otra tambiÃ©n sube).
     - **0** indica que no hay ninguna relaciÃ³n lineal entre las variables.
     - **-1** indica una correlaciÃ³n negativa perfecta (cuando una variable sube, la otra baja).
   - El cÃ³digo `.corr()['knight']` selecciona especÃ­ficamente la columna de correlaciones con respecto a `knight`, es decir, calcula cÃ³mo se relaciona la columna `knight` con todas las demÃ¡s columnas del `DataFrame`.
   - Finalmente, `.sort_values(ascending=False)` ordena estas correlaciones en orden **descendente**, mostrando primero las variables mÃ¡s correlacionadas positivamente con `knight`.

### Â¿Por quÃ© convertir los valores a `1` y `0`?

La conversiÃ³n de etiquetas categÃ³ricas como 'Jedi' y 'Sith' a `1` y `0` se realiza para que se puedan aplicar operaciones matemÃ¡ticas y estadÃ­sticas, como la correlaciÃ³n. La correlaciÃ³n solo puede calcularse entre columnas numÃ©ricas, por lo que este paso de preprocesamiento es necesario para realizar este tipo de anÃ¡lisis.
