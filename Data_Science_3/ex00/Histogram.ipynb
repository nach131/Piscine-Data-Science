```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import math
```

# Test Knight

```python
data = pd.read_csv("subject/Test_knight.csv")
data.head()
```

```python
# Numero de bins y columnas totales del DataFrame
n_bins = 40
num_columns = len(data.columns)

# Definir el nÃºmero de columnas por fila
cols_per_row = 5

# Calcular el nÃºmero total de filas necesarias (redondeando hacia arriba)
num_rows = math.ceil(num_columns / cols_per_row)

# Crear una figura y una cuadrÃ­cula de subplots usando plt.subplots()
fig, axes = plt.subplots(num_rows, cols_per_row, figsize=(25, 25))
fig.tight_layout(pad=5.0)  # Ajustar espacio entre grÃ¡ficos

# Iterar por cada columna del DataFrame y crear su respectivo histograma
for index, column in enumerate(data.columns):
    # Calcular la posiciÃ³n en la cuadrÃ­cula
    row = index // cols_per_row
    col = index % cols_per_row

    # Crear el histograma en la posiciÃ³n correspondiente
    ax = axes[row, col]
    ax.hist(data[column], bins=n_bins, color='#7fbf7f', label='Knight')
    ax.set_title(column)
    ax.legend(loc="upper right")

    # Ajustar el eje y para que los ticks vayan de 5 en 5, comenzando desde 0
    max_y_value = ax.get_ylim()[1]  # Obtener el valor mÃ¡ximo actual del eje y
    ax.set_yticks(range(0, int(max_y_value) + 5, 5))  # Definir los ticks de 5 en 5

# Eliminar grÃ¡ficos vacÃ­os si hay mÃ¡s subplots que columnas en el DataFrame
for i in range(index + 1, num_rows * cols_per_row):
    fig.delaxes(axes.flat[i])


plt.show()
```

# Train knight

```python
data_train = pd.read_csv("subject/Train_knight.csv")
data_train.head()
```

```python
# Numero de bins y columnas totales del DataFrame
n_bins = 40
num_columns = len(data_train.columns)

# Definir el nÃºmero de columnas por fila
cols_per_row = 5

# Calcular el nÃºmero total de filas necesarias (redondeando hacia arriba)
num_rows = math.ceil(num_columns / cols_per_row)

# Crear una figura y una cuadrÃ­cula de subplots usando plt.subplots()
fig, axes = plt.subplots(num_rows, cols_per_row, figsize=(25, 25))
fig.tight_layout(pad=5.0)  # Ajustar espacio entre grÃ¡ficos

# Agrupar los datos por 'knight' y obtener un DataFrame por cada grupo
knights = data_train.groupby('knight')

for index, column in enumerate(data.columns):
    # Calcular la posiciÃ³n en la cuadrÃ­cula
    row = index // cols_per_row
    col = index % cols_per_row

    # Crear el histograma en la posiciÃ³n correspondiente
    ax = axes[row, col]

    if 'Sith' in knights.groups:
        data_sith = knights.get_group('Sith')[column]
        ax.hist(data_sith, bins=n_bins, color='red', alpha=0.5, label='Sith')
    if 'Jedi' in knights.groups:
        data_sith = knights.get_group('Jedi')[column]
        ax.hist(data_sith, bins=n_bins, color='blue', alpha=0.5, label='Jedi')
        
    # Ajustar el tÃ­tulo y la leyenda
    ax.set_title(column)
    ax.legend(loc="upper right")
    
    # Ajustar el eje y para que los ticks vayan de 5 en 5, comenzando desde 0
    max_y_value = ax.get_ylim()[1]  # Obtener el valor mÃ¡ximo actual del eje y
    ax.set_yticks(range(0, int(max_y_value) + 5, 5))  # Definir los ticks de 5 en 5

# Eliminar grÃ¡ficos vacÃ­os si hay mÃ¡s subplots que columnas en el DataFrame
for i in range(index + 1, num_rows * cols_per_row):
    fig.delaxes(axes.flat[i])
    
plt.show()
```
