```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn.metrics import confusion_matrix

def make_matrix(y_true, y_pred):
    # Calcular la matriz de confusión
# Calcular la matriz de confusión usando scikit-learn
    conf_matrix = confusion_matrix(y_true, y_pred)

    # Imprimir la matriz de confusión
    print(f"TP: {TP}, TN: {TN}, FP: {FP}, FN: {FN}")

    # Mostrar en forma de tabla
    conf_matrix = np.array([[TN, FP],[FN, TP]])
    
    print("Matriz de confusión:")
    print(conf_matrix)

    # Graficar la matriz de confusión usando seaborn
    plt.figure(figsize=(6,6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred: 0', 'Pred: 1'], yticklabels=['Real: 0', 'Real: 1'])
    plt.xlabel('Predicción')
    plt.ylabel('Valor Real')
    plt.title('Matriz de Confusión')
    plt.show()
```

### ¿Qué es una Matriz de Confusión?

La **Matriz de Confusión** es una tabla que se utiliza para evaluar el rendimiento de un modelo de clasificación. Te permite comparar las predicciones de un modelo con los valores reales, clasificando los resultados en cuatro categorías:

- **TP** (True Positives o Verdaderos Positivos): predicciones correctas de la clase positiva.
- **TN** (True Negatives o Verdaderos Negativos): predicciones correctas de la clase negativa.
- **FP** (False Positives o Falsos Positivos): predicciones incorrectas de la clase positiva (el modelo predice positivo, pero es negativo).
- **FN** (False Negatives o Falsos Negativos): predicciones incorrectas de la clase negativa (el modelo predice negativo, pero es positivo).

La matriz tiene la siguiente estructura para un problema binario:

|                | Predicción: Negativo | Predicción: Positivo |
|----------------|----------------------|----------------------|
| **Real: Negativo** | TN                   | FP                   |
| **Real: Positivo** | FN                   | TP                   |

### Porcentaje de precisión

Es importante tener en cuenta que los valores de la matriz de confusión se utilizan para calcular métricas de rendimiento importantes como:
- **Precisión** (Accuracy): \((TP + TN) / (TP + TN + FP + FN)\)
- **Recall** (Sensibilidad): \(TP / (TP + FN)\)
- **Precisión** (Precision): \(TP / (TP + FP)\)
